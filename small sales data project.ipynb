# Work of data analyst 

# Data load 

```python
import pandas as pd

# Sales data (CSV ya Excel se load karte hain)
data = {
    "Order_ID": [1001, 1002, 1003, 1004],
    "Date": ["2024-01-05", "2024-01-07", "2024-01-08", "2024-01-08"],
    "Customer": ["Aakash", "Neha", "Rohan", "Jyoti"],
    "Product": ["Mobile", "Shoes", "Laptop", "Saree"],
    "Category": ["Electronics", "Fashion", "Electronics", "Fashion"],
    "Quantity": [1, 2, 1, 3],
    "Price": [15000, 2000, 55000, 1200],
    "Region": ["North", "South", "West", "East"]
}

df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])   # Date format sahi karna
df["Sales"] = df["Quantity"] * df["Price"]  # Naya column Sales
print(df)
```

# Total sales 

```python
total_sales = df["Sales"].sum()
print("Total Sales:", total_sales)
```

# Category wise sales 

```python
category_sales = df.groupby("Category")["Sales"].sum()
print(category_sales)
```

# Region wise sales

```python
region_sales = df.groupby("Region")["Sales"].sum()
print(region_sales)
```

# Monthly sales trend

```python
df["Month"] = df["Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Sales"].sum()
print(monthly_sales)
```

# Visualiaztion 

```python
import matplotlib.pyplot as plt

# Category-wise Sales
category_sales.plot(kind="bar", title="Category-wise Sales")
plt.show()

# Region-wise Sales
region_sales.plot(kind="pie", autopct="%1.1f%%", title="Region-wise Sales")
plt.show()

# Monthly Sales Trend
monthly_sales.plot(kind="line", marker="o", title="Monthly Sales Trend")
plt.show()
```
